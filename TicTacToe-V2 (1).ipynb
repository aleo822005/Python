```python

def sample_tbl():
    print(sample[0:3])
    print(sample[3:6])
    print(sample[6:9])
def user_play():
    #display(main_board)
    sample_tbl()
    ps=input("Please enter the position you wish to mark 'X' (1-9)")
            
    #ps=int(ps)
    #ps=ps-1
    
    if ps not in ['1','2','3','4','5','6','7','8','9'] or ps=='':
        clear_output()
        print("'"+str(ps)+"' is not a valid input.Please try again")
        user_play()
    elif main_board[int(ps)-1]=='X' or main_board[int(ps)-1]=='O':
        clear_output()
        print("That position is taken; please choose another")
        user_play()
    else:
        main_board[int(ps)-1]='X'
        board_score[int(ps)-1]='X'
        sample[int(ps)-1]=0
        split_rows()
        board_score
        #mech_play(board_score)
    display(main_board)
    #display(board_score)

def Game_full():
    blanks=0
    for t in main_board:
        if t==' ':
            blanks+=1
    if blanks==0 or blanks==1:
        return True
    else:
        return False
def check_win(mark):
    if (main_board[0]==main_board[1]==main_board[2]==mark):
        return True
    elif (main_board[3]==main_board[4]==main_board[5]==mark):
        return True
    elif (main_board[6]==main_board[7]==main_board[8]==mark):
        return True
    elif (main_board[0]==main_board[4]==main_board[8]==mark):
        return True
    elif (main_board[6]==main_board[4]==main_board[2]==mark):
        return True
    elif (main_board[0]==main_board[3]==main_board[6]==mark):
        return True
    elif (main_board[1]==main_board[4]==main_board[7]==mark):
        return True
    elif (main_board[2]==main_board[5]==main_board[8]==mark):
        return True
    else:
        return False
            
def display(board):
    clear_output()
    
    print(' -------------')
    print(' | '+str(board[0])+' | '+str(board[1])+' | '+str(board[2])+' | ')
    print(' -------------')
    print(' | '+str(board[3])+' | '+str(board[4])+' | '+str(board[5])+' | ')
    print(' -------------')
    print(' | '+str(board[6])+' | '+str(board[7])+' | '+str(board[8])+' | ')
    print(' -------------')
def split_rows():
    a=board_score[0:3]
    x_score(a)
    b=board_score[3:6]
    x_score(b)
    c=board_score[6:9]
    x_score(c)
    #d=board_score[0]+','+board_score[4]+board_score[8]
    d=[]
    d.append(board_score[0])
    d.append(board_score[4])
    d.append(board_score[8])
    x_score(d)
    e=[]
    e.append(board_score[6])
    e.append(board_score[4])
    e.append(board_score[2])
    x_score(e)
    f=[]
    f.append(board_score[0])
    f.append(board_score[3])
    f.append(board_score[6])
    x_score(f)
    g=[]
    g.append(board_score[1])
    g.append(board_score[4])
    g.append(board_score[7])
    x_score(g)
    h=[]
    h.append(board_score[2])
    h.append(board_score[5])
    h.append(board_score[8])
    x_score(h)
    recombine_rows(a,b,c,d,e,f,g,h,board_score)
    return a,b,c,d,e,f,g,h,board_score

def x_score(lst):
    
    wt=0
    wt_O=0
    flag=0
    for a in range(len(lst)):
        if lst[a] =='X':
            wt=wt+1
        if lst[a] =='O':
            wt_O=wt_O+1
    if wt!=0 and wt_O!=0:flag=2
    for a in range(len(lst)):
        if lst[a]!='X' and lst[a]!='O':
            if flag!=2 and wt!=2 and wt_O!=2:
                lst[a]=lst[a]+wt
            elif wt==2:
                lst[a]=100
            elif wt_O==2:
                lst[a]=500
            else:
                lst[a]=-1
    
    return lst

def recombine_rows(a,b,c,d,e,f,g,h,board_score):
    if board_score[0]!='X' and board_score[0]!='O':board_score[0]=round((a[0]+d[0]+f[0])/3,1)
    if board_score[1]!='X' and board_score[1]!='O':board_score[1]=round((a[1]+g[0])/2,1)
    if board_score[2]!='X' and board_score[2]!='O':board_score[2]=round((a[2]+e[2]+h[0])/3,1)
    
    if board_score[3]!='X' and board_score[3]!='O':board_score[3]=round((b[0]+f[1])/2,1)
    if board_score[4]!='X' and board_score[4]!='O':board_score[4]=round((b[1]+d[1]+e[1]+g[1])/4,1)
    if board_score[5]!='X' and board_score[5]!='O':board_score[5]=round((b[2]+h[1])/2,1)
    
    if board_score[6]!='X' and board_score[6]!='O':board_score[6]=round((c[0]+e[0]+f[2])/3,1)
    if board_score[7]!='X' and board_score[7]!='O':board_score[7]=round((c[1]+g[2])/2,1)
    if board_score[8]!='X' and board_score[8]!='O':board_score[8]=round((c[2]+d[2]+h[2])/3,1)
    
    return board_score

def mech_play(board_score):
    
    board_score2=[]
    
    
    for i in range(len(board_score)):
        if board_score[i]=='X' or board_score[i]=='O':
            board_score2.append(-2)
        else:
            board_score2.append(board_score[i])
            
    
    
    next_O=board_score2.index(max(board_score2))
    
    main_board[next_O]='O'
    board_score[next_O]='O'
    sample[next_O]=0
    
    return main_board

```
